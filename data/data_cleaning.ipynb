```python
import numpy as np
import pandas as pd
```

```python
games = pd.read_csv('games.csv')
clubs = pd.read_csv('clubs.csv')
```

```python
# count the unique club_ids in clubs
print(clubs.club_id.nunique())
```

```python
clubs
```

```python
# replace the name of the clubs where the club name is null
# use modified club_code
clubs['name'] = clubs.apply(
    lambda row: row['club_code'].replace('-', ' ') if pd.isna(row['name']) or row['name'] == '' else row['name'], 
    axis=1
)
```

```python
clubs
```

```python
clubs.to_csv('clubs_cleaned.csv', index=False)
```

```python
games
```

```python
# if home_club_id is not in clubs, then drop the row
games = games[games.home_club_id.isin(clubs.club_id)]
games = games[games.away_club_id.isin(clubs.club_id)]
games
```

```python
games.to_csv('games_cleaned.csv', index=False)
```

```python
appearances = pd.read_csv('appearances.csv')
appearances
```

```python
# if game_id is not in games, then drop the row
appearances = appearances[appearances.game_id.isin(games.game_id)]
appearances
```

```python
appearances.to_csv('appearances_cleaned.csv', index=False)
```

```python
club_games = pd.read_csv('club_games.csv')
club_games
```

```python
# if game_id is not in games, then drop the row
club_games = club_games[club_games.game_id.isin(games.game_id)]
club_games
```

```python
club_games.to_csv('club_games_cleaned.csv', index=False)
```

```python
players = pd.read_csv('players.csv')
players
```

```python
# if player_id is not in players, then drop the row from appearances
appearances = appearances[appearances.player_id.isin(players.player_id)]
appearances
```

```python
appearances.to_csv('appearances_cleaned.csv', index=False)
```
